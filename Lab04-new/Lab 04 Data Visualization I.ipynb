```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

## A. Basic Plotting I

```python
months = ['Jan', 'Feb', 'March', 'April', 'May', 'June']
monthly_sales = [13, 5, 7, 14, 10, 12]

plt.figure(figsize=(6, 4))
plt.plot(months, 
         monthly_sales, 
         marker='o', 
         markersize=10,
         markerfacecolor="None",
         linewidth=2, 
         color='#6bd9d9',
         clip_on=False
        )
plt.title('Print Sales for January to June, 2022', size=16)
plt.xlabel('Month', size=15)
plt.ylabel('Monthly Sales ($1000)', size=15)
plt.xlim(months[0], months[-1])
plt.xticks(months)
plt.yticks([0, 5, 10, 15])

plt.savefig('./Lab04_A.png', dpi=400)
plt.show()
```

## B. Basic Plotting II

```python
x = np.linspace(0, 0.5, 50)
y1 = np.sin(2*np.pi*5*x)
y2 = np.sin(2*np.pi*5*x + np.pi/2)

plt.figure(figsize=(6, 4))
plt.plot(x, 
         y1, 
         '*-', 
         color='#02bfbf',
         linewidth=0.8,
         clip_on=False
        )
plt.plot(x, 
         y2,
         'o:k',
         mfc='None',
         linewidth=0.8,
         clip_on=False
        )
plt.xlim(0, 0.5)
plt.yticks(np.linspace(-1.5, 1.5, 7))
plt.ylim(-1.5, 1.5)
plt.title('In-Phase $\it{(solid)}$ and Quadarture $\it(dotted)$ Signals', size=14)
plt.xlabel('Time (μs)', size=13, fontweight='bold')
plt.ylabel("$\it{'Normalized'}$ Signals", size=13)
plt.text(0.19, 1.25, r'$\frac{π}{2}$ phase lag', size=8.5)
plt.arrow(0.21, 1.15, 0.032, 0, 
          head_width=0.05,
          head_length=0.005, 
          overhang=2,
          fc='k', 
          ec='k')
plt.arrow(0.24, 1.15, -0.031, 0,
          head_width=0.05, 
          head_length=0.005,
          overhang=2, 
          fc='k', 
          ec='k')

plt.savefig('./Lab04_B.png', dpi=400)
plt.show()
```

## C. Basic Plotting III 

```python
x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x) + np.sin(3*x)

fig, ax = plt.subplots()
ax.plot(x, y, color='magenta', linewidth=3)
ax.set_xlabel('x-axis', color='#01ff01')
ax.set_xticks(np.linspace(0, 14, 8), color='#01ff01')
ax.set_xlim(0, 14)
ax.set_yticks([-1, -0.3, 0.1, 1], ['Minimum', 'Critical', 'Collapse', 'maximum'])
ax.grid()
ax.tick_params(top=True, right=True, direction='in')
ax.tick_params(axis='x', colors='#01ff01', width=2, grid_color='#01ff01')
ax.tick_params(axis='y', width=1.5, length=6, grid_color='black', grid_linewidth=0.4)
ax.spines['top'].set(color='#01ff01', linewidth=1.5)
ax.spines['bottom'].set(color='#01ff01', linewidth=1.5)
ax.spines['left'].set(linewidth=1.5)

plt.savefig('./Lab04_C.png', dpi=400)
plt.show()
```

## D. Scatter Plot

```python
df = pd.read_csv('./pokemon_data.csv')
cluster_zero = df.loc[df["cluster"] == 0]
cluster_one = df.loc[df["cluster"] == 1]
cluster_two = df.loc[df["cluster"] == 2]

centroids_x = [49.875000, 79.801887, 123.270833]
centroids_y = [48.075000, 74.386792, 102.479167]
colors = ['#bf00bf', '#008000', '#00bfbf']
labels = ['Cent. of cluster0', 'Cent. of cluster1', 'Cent of cluster2']

fig, ax = plt.subplots(1, 1, figsize=(9, 8))

ax.scatter(cluster_zero.Attack, cluster_zero.Defense,
           marker='.',
           color='magenta',
           alpha=0.3,
           label='Cluster 0'
          )
ax.scatter(cluster_one.Attack, cluster_one.Defense,
           marker='.',
           color='green',
           alpha=0.3,
           label='Cluster 1'
          )
ax.scatter(cluster_two.Attack, cluster_two.Defense,
           marker='.',
           color='cyan',
           alpha=0.3,
           label='Cluster 2'
          )
ax.scatter(centroids_x[0], centroids_y[0], 
           marker='^',
           color=colors[0],
           label='Cent. of cluster0'
          )
ax.scatter(centroids_x[1], centroids_y[1], 
           marker='^',
           color=colors[1],
           label='Cent. of cluster1'
          )
ax.scatter(centroids_x[2], centroids_y[2], 
           marker='^',
           color=colors[2],
           label='Cent. of cluster2'
          )
ax.set_title('Scatter plot of pokemons')
ax.set_xlabel('Attack')
ax.set_ylabel('Defense')

plt.legend(ncol=2)
plt.savefig('./Lab04_D.png', dpi=400)
plt.show()
```
